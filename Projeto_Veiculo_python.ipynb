```python
from abc import ABC, abstractmethod

class Veiculo(ABC):
    def __init__(self, marca, modelo, ano):
        self._marca = marca
        self._modelo = modelo
        self._ano = ano

    @abstractmethod
    def ligar(self):
        pass

    @abstractmethod
    def mover(self):
        pass

    def exibir_info(self):
        print(f"{self._marca} {self._modelo} - Ano: {self._ano}")


class Automovel(Veiculo):
    def mover(self):
        print(f"{self._modelo} está dirigindo na estrada.")

class Motocicleta(Veiculo):
    def mover(self):
        print(f"{self._modelo} está acelerando na pista.")

class Bicicleta(Veiculo):
    def ligar(self):
        print("Bicicleta não precisa ligar (a menos que seja elétrica).")

    def mover(self):
        print(f"{self._modelo} está pedalando.")

class VeiculoAquatico(Veiculo):
    def mover(self):
        print(f"{self._modelo} está navegando.")

class VeiculoAereo(Veiculo):
    def mover(self):
        print(f"{self._modelo} está voando.")


class Carro(Automovel):
    def ligar(self):
        print("Ligando carro com chave ou botão.")

class Caminhao(Automovel):
    def ligar(self):
        print("Ligando caminhão com ignição pesada.")

class Onibus(Automovel):
    def ligar(self):
        print("Ligando ônibus para transporte coletivo.")

class SUV(Automovel):
    def ligar(self):
        print("Ligando SUV com tração nas quatro rodas.")

class Conversivel(Automovel):
    def ligar(self):
        print("Ligando conversível elegante.")


class Scooter(Motocicleta):
    def ligar(self):
        print("Scooter ligando com botão eletrônico.")

class Esportiva(Motocicleta):
    def ligar(self):
        print("Moto esportiva roncando alto.")

class Custom(Motocicleta):
    def ligar(self):
        print("Moto custom aquecendo motor.")

class Naked(Motocicleta):
    def ligar(self):
        print("Moto naked pronta para o uso urbano.")

class Trail(Motocicleta):
    def ligar(self):
        print("Moto trail ligando para trilha.")


class Urbana(Bicicleta):
    pass

class Montanha(Bicicleta):
    pass

class Eletrica(Bicicleta):
    def ligar(self):
        print("Bicicleta elétrica sendo ativada.")

class Dobravavel(Bicicleta):
    pass

class Speed(Bicicleta):
    pass


class Lancha(VeiculoAquatico):
    def ligar(self):
        print("Ligando lancha potente.")

class Iate(VeiculoAquatico):
    def ligar(self):
        print("Iate de luxo iniciando.")

class JetSki(VeiculoAquatico):
    def ligar(self):
        print("Jet Ski rugindo nas águas.")

class Submarino(VeiculoAquatico):
    def ligar(self):
        print("Submarino iniciando submersão.")

class BarcoPesca(VeiculoAquatico):
    def ligar(self):
        print("Barco de pesca ligando para jornada.")


class Aviao(VeiculoAereo):
    def ligar(self):
        print("Avião ligando turbinas.")

class Helicoptero(VeiculoAereo):
    def ligar(self):
        print("Hélice do helicóptero girando.")

class Drone(VeiculoAereo):
    def ligar(self):
        print("Drone decolando.")

class Planador(VeiculoAereo):
    def ligar(self):
        print("Planador puxado para decolagem.")

class Balao(VeiculoAereo):
    def ligar(self):
        print("Balão sendo aquecido.")


def main():
    veiculos = [
        Carro("Toyota", "Corolla", 2020),
        Caminhao("Volvo", "FH16", 2019),
        Onibus("Mercedes", "Torino", 2018),
        SUV("Jeep", "Renegade", 2021),
        Conversivel("BMW", "Z4", 2022),

        Scooter("Honda", "PCX", 2021),
        Esportiva("Yamaha", "R1", 2022),
        Custom("Harley", "Iron 883", 2020),
        Naked("Kawasaki", "Z400", 2021),
        Trail("Honda", "XRE 300", 2020),

        Urbana("Caloi", "Urbana 500", 2019),
        Montanha("Soul", "SL329", 2021),
        Eletrica("Lev", "E-bike", 2022),
        Dobravavel("Tern", "Link D8", 2020),
        Speed("Specialized", "Allez", 2021),

        Lancha("Yamaha", "FX Cruiser", 2018),
        Iate("Azimut", "Grande 27M", 2022),
        JetSki("Sea-Doo", "GTX 230", 2021),
        Submarino("OceanGate", "Cyclops", 2017),
        BarcoPesca("Mariner", "Pescador 300", 2019),

        Aviao("Boeing", "737", 2015),
        Helicoptero("Airbus", "H130", 2020),
        Drone("DJI", "Mavic Air 2", 2021),
        Planador("Schempp-Hirth", "Duo Discus", 2018),
        Balao("Cameron", "Z-105", 2016),

        Carro("Fiat", "Uno", 2010),
        Esportiva("Suzuki", "Bandit", 2005),
        Caminhao("Scania", "P360", 2017),
        Onibus("Volare", "V8L", 2016),
        SUV("Toyota", "SW4", 2022)
    ]

    print("--- Exemplos de Veículos ---")
    print("-" * 40)
    for veiculo in veiculos:
        veiculo.exibir_info()
        veiculo.ligar()
        veiculo.mover()
        print("-" * 40)

if __name__ == "__main__":
    main()
```
